```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# (a) Initialize gradient descent function
def gradient_descent(x0, y0, grad_f, alpha, num_iterations, tolerance=1e-6):
    x, y = x0, y0 
    
    for _ in range(num_iterations):
        # (b) Compute the gradient
        grad_x, grad_y = grad_f(x, y)
        
        # (c) Update variables
        new_x = x - alpha * grad_x
        new_y = y - alpha * grad_y
        
        # (d) Check for convergence
        if abs(new_x - x) < tolerance and abs(new_y - y) < tolerance:
            break
        
        x, y = new_x, new_y
    
    return x, y

# Function 1: Quadratic function and its gradient
def fun_1(x, y):
    return x**2 + y**2

def grad_f_1(x, y):
    return 2*x, 2*y

# Function 2: Exponential-based function and its gradient
def fun_2(x, y):
    return 1 - np.exp(-x**2 - (y-2)**2) - 2*np.exp(-x**2 - (y+2)**2)

def grad_f_2(x, y):
    exp1 = np.exp(-x**2 - (y-2)**2)
    exp2 = np.exp(-x**2 - (y+2)**2)
    grad_x = 2*x * (exp1 + 2*exp2)
    grad_y = 2*(y-2) * exp1 + 4*(y+2) * exp2
    return grad_x, grad_y

# Function to plot 3D surfaces
def plot_function(f, title):
    """
    Plots a 3D surface of a given function.
    """
    X, Y = np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)
    x, y = np.meshgrid(X, Y)
    z = f(x, y)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
    ax.set_title(title)
    plt.show()
```

```python
# Running gradient descent on Function 1
result_1a = gradient_descent(0.1, 0.1, grad_f_1, 0.1, 10)
result_1b = gradient_descent(-1, 1, grad_f_1, 0.01, 100)

print("Gradient Descent Results for Function 1 (x^2 + y^2):")
print("Start at (0.1, 0.1) -> Minimum at", result_1a)
print("Start at (-1, 1) -> Minimum at", result_1b)
```

```python
# Running gradient descent on Function 2
result_2a = gradient_descent(0, 1, grad_f_2, 0.01, 1000)
result_2b = gradient_descent(0, -1, grad_f_2, 0.01, 1000)

print("Gradient Descent Results for Function 2 (Complex Exponential Function):")
print("Start at (0,1) -> Minimum at", result_2a)
print("Start at (0,-1) -> Minimum at", result_2b)
```

```python
plot_function(fun_1, "Function 1: x^2 + y^2")
```

```python
plot_function(fun_2, "Function 2: Complex Exponential Function")
```
