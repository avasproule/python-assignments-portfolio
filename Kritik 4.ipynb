```python
import numpy as np
import matplotlib.pyplot as plt

# Function to perform gradient descent and locate the minimum of a given function
def gradient_descent(func, step_size, start_point):
    # Compute the numerical derivative of the function at a specific x value
    def compute_slope(func, at_point):
        return (func(at_point + 1e-10) - func(at_point - 1e-10)) / (2 * 1e-10)

    # Lists to store the x values and their corresponding function values (y values)
    x_vals = [start_point]
    y_vals = [func(start_point)]  # Evaluate the function at the starting x value

    # Perform 1000 iterations of gradient descent
    for iteration in range(1000):
        # Calculate the gradient (slope) at the current x value
        slope = compute_slope(func, x_vals[-1])
        # Adjust x by moving in the direction opposite to the slope
        delta_x = -step_size * slope
        x_vals.append(x_vals[-1] + delta_x)  # Append the updated x value
        # Evaluate the function at the new x value and store it
        y_vals.append(func(x_vals[-1]))

    # Prepare data to plot the function curve
    x_range = np.linspace(min(x_vals) - 0.5, max(x_vals) + 0.5, 1000)
    y_range = [func(i) for i in x_range]

    # Create the plot
    plt.plot(x_range, y_range, label='Function f(x)')
    plt.plot(x_vals, y_vals, marker='o', color='r', label='Steps of Gradient Descent')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.title('Finding the Minimum Using Gradient Descent')
    plt.legend()
    plt.show()

    # Return the final x value (approximation of the minimum) and its function value
    return round(x_vals[-1], 3), round(y_vals[-1], 3)

# Define the function we are trying to minimize
def func(x):
    if x > 0:
        return x ** x
    elif x == 0:
        return 1
    else:
        return abs(x) ** abs(x)

# Set the learning rate and the initial x value
step_size = 0.01
start_point = 1.5

# Run the gradient descent function and display the results
min_point, min_value = gradient_descent(func, step_size, start_point)
print("Optimal X:", min_point)  # x value of the minimum point
print("Optimal Y:", min_value)  # smallest function value
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to perform gradient descent to find the minimum of a function
def gradient_descent(func, step_size, start_point):
    # Function to compute the numerical derivative at a given x value
    def compute_slope(func, at_point):
        return (func(at_point + 1e-10) - func(at_point - 1e-10)) / (2 * 1e-10)

    # Lists to store the x values and their corresponding function values (y values)
    x_vals = [start_point]
    y_vals = [func(start_point)]  # Evaluate the function at the starting x value

    # Perform 1000 steps of gradient descent
    for iteration in range(1000):
        # Calculate the gradient (slope) at the current x value
        slope = compute_slope(func, x_vals[-1])
        # Update x by moving opposite to the slope
        delta_x = -step_size * slope
        x_vals.append(x_vals[-1] + delta_x)  # Append the new x value
        # Evaluate the function at the new x value and store it
        y_vals.append(func(x_vals[-1]))

    # Prepare the range of x values to plot the function curve
    x_range = np.linspace(min(x_vals) - 0.5, max(x_vals) + 0.5, 1000)
    y_range = [func(i) for i in x_range]

    # Plot the function and gradient descent steps
    plt.plot(x_range, y_range, label='Function |x|')
    plt.plot(x_vals, y_vals, marker='o', color='r', label='Steps of Gradient Descent')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.title('Gradient Descent on |x| (Fails Due to Undefined Derivative at X=0)')
    plt.legend()
    plt.show()

    # Return the final x value (approximation of the minimum) and its function value
    return round(x_vals[-1], 3), round(y_vals[-1], 3)

# Define the absolute value function to minimize
def func(x):
    return abs(x)

# Set the learning rate and starting point
step_size = 0.1
start_point = 1.5

# Attempt to run the gradient descent function and display the results
optimal_x, optimal_y = gradient_descent(func, step_size, start_point)
print("Optimal X:", optimal_x)  # x value of the minimum point
print("Optimal Y:", optimal_y)  # smallest function value

# This method does not work for |x| because the derivative is undefined at x=0.
# Gradient descent relies on the derivative to compute the direction of steepest descent,
# so it cannot handle functions with undefined or discontinuous derivatives at critical points.
```
