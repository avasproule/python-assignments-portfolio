```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define symbolic variables
x, y = sp.symbols('x y')

# Function definitions and differentiation (Section A)
f1 = x**2 + sp.sin(y)
df1_dx = sp.diff(f1, x)
df1_dy = sp.diff(f1, y)

f2 = sp.exp(x) * sp.sin(y) + y**3
df2_dx = sp.diff(f2, x)
df2_dy = sp.diff(f2, y)

f3 = sp.exp(x) * sp.sin(y)
df3_xx = sp.diff(f3, x, x)

f4 = sp.exp(x) * sp.cos(y) + 3*y**2
df4_yy = sp.diff(f4, x)

# Print results
print(df2_dx, df2_dy)
print(df3_xx, df4_yy)
```

```python
# Gradient calculation (Section B)
f_grad = x**2 * y + x + y**2
df_grad_x = sp.diff(f_grad, x)
df_grad_y = sp.diff(f_grad, y)
gradient_magnitude = sp.sqrt(df_grad_x**2 + df_grad_y**2)
print(gradient_magnitude)
```

```python
# Second-order derivatives (Section C)
f_log = sp.log(x**2 + y**2 + 1e-60)
df_log_dx = sp.diff(f_log, x)
df_log_dy = sp.diff(f_log, y)

f_sec_x = 2*x / (x**2 + y**2 + 1e-60)
df_sec_xx = sp.diff(f_sec_x, x)

f_sec_y = 2*y / (x**2 + y**2 + 1e-60)
df_sec_yy = sp.diff(f_sec_y, x)

print(df_log_dx, "and", df_log_dy)
print(df_sec_xx, "and", df_sec_yy)
```

```python
# Contour Plot (Section D)
j1 = x**3 - 3*x*y + y**3
j1_func = sp.lambdify((x, y), j1, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z1 = j1_func(X, Y)

plt.contourf(X, Y, Z1, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
# Another Contour Plot (Section E)
j2 = x**2 - 5*x*y
j2_func = sp.lambdify((x, y), j2, 'numpy')

Z2 = j2_func(X, Y)

plt.contourf(X, Y, Z2, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^2 - 5xy$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
